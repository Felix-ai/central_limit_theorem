```python
# Import Modules
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use("ggplot")
```

##### Central Limit Theorem with a Normal Population

```python
# Parameters
mu = 75
sigma = 8
N = 100000
n = 10
bins = 50
```

```python
# Normal Distribution
x = np.random.normal(mu, sigma, N)
```

```python
# Distribution Plot
fig, ax = plt.subplots(figsize=(12, 8))
ax.hist(x, bins=bins, color="c")
ax.set_xlabel("Weight (Kg)", color="b")
ax.set_ylabel("Frequency", color="b")
ax.set_title("Normal Population")
plt.savefig("images/norm_population.png", bbox_inches="tight")
```

```python
# Sample Means
x_bars = [np.mean(random.choices(x, k=n)) for nn in range(1, N)]
```

```python
# Sample Means Distribution
fig, ax = plt.subplots(figsize=(12, 8))
ax.hist(x_bars, bins=bins, color="c")
ax.set_xlabel("Mean: Samples of n={:.0f}".format(n), color="b")
ax.set_ylabel("Frequency", color="b")
ax.set_title("Sample Means Distribution")
plt.savefig("images/norm_population_sample_means.png", bbox_inches="tight")
```

##### Central Limit Theorem with a Dichotomous Outcome

```python
# Parameters
pop = [1, 0]
index = ["Yes", "No"]
p = 0.3
weights = [p, 1-p]
trials = 1
n = 20
bins = 15
```

```python
# Population
# xd = pd.Series(random.choices(pop, weights=weights, k=N))
xd = pd.Series(np.random.binomial(trials, p, N))
xd_cnts = xd.value_counts().to_frame("cnts")
xd_cnts["prp"] = xd_cnts["cnts"] / xd_cnts["cnts"].sum()
xd_cnts.index = index[::-1]
```

```python
# Distribution Plot
fig, ax = plt.subplots(figsize=(8, 8))
plt.bar(xd_cnts.index, xd_cnts["prp"], color="c")
ax.set_xlabel("Success of a medical procedure", color="b")
ax.set_ylabel("Proportion", color="b")
ax.set_title("Dichotomous Outcome")
plt.savefig("images/dichotomous_outcome.png", bbox_inches="tight")
```

```python
# Sample Means
if min(n*p, n*(1-p)) >= 5:
    xd_bars = [n * np.mean(random.choices(xd, k=n)) for nn in range(1, N)]
```

```python
# Sample Means Distribution
fig, ax = plt.subplots(figsize=(12, 8))
ax.hist(xd_bars, bins, color="c")
ax.set_xlabel("Mean: Samples of n={:.0f}".format(n), color="b")
ax.set_ylabel("Frequency", color="b")
ax.set_title("Sample Means Distribution")
plt.savefig("images/dichotomous_outcome_sample_means.png", bbox_inches="tight")
```

##### Central Limit Theorem with a Skewed Distribution

```python
# Parameters
lam = 3
n = 30
bins = 12
```

```python
# Population
xp = np.random.poisson(lam=lam, size=N)
```

```python
# Distribution Plot
fig, ax = plt.subplots(figsize=(12, 8))
ax.hist(xp, bins=bins, color="c")
ax.set_xlabel("Rates", color="b")
ax.set_ylabel("Frequency", color="b")
ax.set_title("Poisson Distribution")
plt.savefig("images/poisson_population.png", bbox_inches="tight")
```

```python
# Sample Means
xp_bars = [np.mean(random.choices(xp, k=n)) for nn in range(1, N)]
```

```python
# Sample Means Distribution
fig, ax = plt.subplots(figsize=(12, 8))
ax.hist(xp_bars, bins, color="c")
ax.set_xlabel("Mean: Samples of n={:.0f}".format(n), color="b")
ax.set_ylabel("Frequency", color="b")
ax.set_title("Sample Means Distribution")
plt.savefig("images/poisson_population_sample_means.png", bbox_inches="tight")
```
